```python
import json
import pandas as pd
import seaborn as sns

from pathlib import Path
```

```python
with open("Results/metrics.json") as f:
    metrics = json.load(f)
```

```python
prettier = {}
for k,v in metrics["satellite"].items():
    prettier[Path(k).stem] = {
        "mse x 100": v["mse"]*100,
        "l1 x 10": v["l1"]*10,
        "ssim": v["ssim"],
    }
```

```python
df = pd.DataFrame(prettier).transpose()
sns.heatmap(df, annot=True)
```
